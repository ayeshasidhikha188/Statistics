## Matplotlib Data visualization

```python
import matplotlib.pyplot as plt ,numpy as np
```

```python
# straight line 
x=np.linspace(0,10,10)
```

```python
x
```

```python
plt.plot(x)# x and y axis are same 
```

```python
plt.plot(x)
plt.plot(-x)# reverse# multiply '-' for entire array  
```

```python
# color changing 
```

```python
plt.plot(x,color='green')
plt.plot(-x,color='red')# reverse# multiply '-' for entire array 
```

```python
# increasing x limit 
```

```python
# f(x)=x,f(x)=-x
# y will be the function of x
```

```python
plt.plot(x,color='green')
plt.plot(-x,color='red')
plt.xlim(0,25)
```

```python
plt.plot(x,color='green')
plt.plot(x**2,color='red')
plt.xlim(0,10)
```

```python
# multiple plots on the same canvas 
 # on the same plot can draw multiple plots
```

```python
plt.plot(x,color='green')
plt.plot(x**2,color='red')
plt.plot(5*x+2,color='blue')
plt.xlim(0,10)
```

```python
plt.plot(x,color='green')
plt.plot(x**2,color='red')
plt.plot(5*x+2,color='blue')
plt.xlim(-10,10)
```

```python
# creating variable 
f=plt.figure() # canvas and axis provides
#<Figure size 640x480 with 0 Axes>
# what ever we plot it will be in same figure
ax1=f.add_axes(rect=[0.5,0.0,0.5,0.5])# add axis default it is rectangle 
# entire canvas 
# left,bottom,height,top 
```

```python
f=plt.figure()
ax1=f.add_axes(rect=[0.5,0.0,0.5,0.5])
ax1.plot(x)
plt.show()
```

```python
# adding another axis 
#ax2
f=plt.figure()
ax1=f.add_axes(rect=[0.5,0.0,0.5,0.5])# left,bottom, top, height
ax2=f.add_axes(rect=[1.0,0.5,1.0,0.5])# left, bottom, top,height 
ax2.plot(-x)
ax1.plot(x)
plt.show()
```

```python
f=plt.figure()
ax1=f.add_axes(rect=[0.5,0.0,0.5,0.5])# left,bottom, top, height
ax2=f.add_axes(rect=[1.0,0.5,1.0,0.5])# left, bottom, top,height
ax3=f.add_axes(rect=[1.0,0.0,0.5,0.5])
ax4=f.add_axes(rect=[0.0,0.5,1.0,0.5])
ax2.plot(-x)
ax1.plot(x)
ax3.plot(-x)
ax4.plot(x)
plt.show()
```
