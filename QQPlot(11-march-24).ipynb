```python
## normal distribution
#'Normal':np.random.normal(0,1,100)mean=0,standdeviation=1
df=pd.DataFrame({'Normal':np.random.normal(0,1,100),
               'Rskew':np.random.exponential(1.0,100),# 1.0 is sd
               'Lskew':-np.random.exponential(1.0,100)+3,
               'Uniform':np.random.uniform(0,1,100),
               'Pareto':np.random.pareto(3,100),
               'Power':np.random.power(3,100)}) 
```

```python
### QQPlots for 6 distributions 
# by default it is normal distribution 
# comapiring with normal distribution
```

```python
plt.figure(figsize=(10,30))
fig,ax=plt.subplots(nrows=2,ncols=3,figsize=(30,20))
qqplot(df['Normal'],line='45',ax=ax[0,0])
qqplot(df['Rskew'],line='45',ax=ax[0,1])
qqplot(df['Lskew'],line='45',ax=ax[0,2])
qqplot(df['Uniform'],line='45',ax=ax[1,0])
qqplot(df['Pareto'],line='45',ax=ax[1,1])
qqplot(df['Power'],line='45',ax=ax[1,2])

plt.show() 
```

```python
help(smapiQqplot)
```

```python
plt.figure(figsize=(10,30))
fig,ax=plt.subplots(nrows=2,ncols=3,figsize=(30,20))
qqplot(df['Normal'],line='r',ax=ax[0,0])
qqplot(df['Rskew'],line='r',ax=ax[0,1])
qqplot(df['Lskew'],line='r',ax=ax[0,2])
qqplot(df['Uniform'],line='r',ax=ax[1,0])
qqplot(df['Pareto'],line='r',ax=ax[1,1])
qqplot(df['Power'],line='r',ax=ax[1,2])

plt.show() 
```

```python
plt.figure(figsize=(30,3))
fig,ax=plt.subplots(nrows=2,ncols=3,figsize=(30,15))
qqplot(df['Normal'],dist=scipy.stats.expon,line='45',ax=ax[0,0])
qqplot(df['Rskew'],dist=scipy.stats.expon,line='45',ax=ax[0,1])
qqplot(df['Lskew'],dist=scipy.stats.expon,line='45',ax=ax[0,2])
qqplot(df['Uniform'],dist=scipy.stats.expon,line='45',ax=ax[1,0])
qqplot(df['Pareto'],dist=scipy.stats.expon,line='45',ax=ax[1,1])
qqplot(df['Power'],dist=scipy.stats.expon,line='45',ax=ax[1,2])
plt.show() 
```
